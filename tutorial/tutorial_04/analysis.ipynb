```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('white')

from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("v_CO.csv")
x = df["sigma"]
y = df["v_CO_asy"]

f = plt.figure(figsize=(8,8))
plt.scatter(x, y, c="black", label="")

m, b = np.polyfit(x, y, 1)
pred = x * m + b
r2 = r2_score(y, pred)

plt.plot(x, pred, "-", color="black", alpha=0.3)

plt.annotate(f"vCO = {m:.2f} * σp + {b:.2f}\nR2 = {r2:.3f}", (0.2, 2077), ha='left', fontsize=12) 

plt.ylabel("CO Stretching Frequency (cm-1)")
plt.xlabel("σp")
plt.title("CO Stretching Frequency Depends on Ligand Electronic Parameters", fontsize="12", fontweight="bold")
plt.show()
```
