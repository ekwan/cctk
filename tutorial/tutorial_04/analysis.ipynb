```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style("white")
```

```python
df = pd.read_csv("output.csv", sep=",")
df.columns = ["file", "energy", "rel_energy", "iter", "mul_q", "hir_q", "success", "imag", "dist"]
df.describe()

otf_energy = df.loc[df["file"] == "OTf_anion.out"]["energy"].values[0]
sbf6_energy = df.loc[df["file"] == "SbF6_anion.out"]["energy"].values[0]
cu2_energy = df.loc[df["file"] == "Cu2-tBu2BOX-dication.out"]["energy"].values[0]


df["pairing_energy"] = 0
df.loc[df["file"].str.contains("OTf_c"), "pairing_energy"] = (df.loc[df["file"].str.contains("OTf_c"), "energy"] - otf_energy - cu2_energy) * 627.509
df.loc[df["file"].str.contains("SbF6_c"), "pairing_energy"] = (df.loc[df["file"].str.contains("SbF6_c"), "energy"] - sbf6_energy - cu2_energy) * 627.509
df.head(40)
```

```python
fig = plt.figure(figsize=(8,8))
plt.scatter(df.loc[df["file"].str.contains("OTf_c"), "pairing_energy"],df.loc[df["file"].str.contains("OTf_c"), "dist"], color="red", label="OTf")
plt.scatter(df.loc[df["file"].str.contains("SbF6_c"), "pairing_energy"],df.loc[df["file"].str.contains("SbF6_c"), "dist"], color="blue", label="SbF6")
plt.legend(loc="best")
plt.ylabel("Cation–Anion Distance (Å)")
plt.xlabel("Ion-Pair Stabilization (kcal/mol)")
plt.title("Ion-Pair Energy Correlates with Separation", fontsize="12", fontweight="bold")
plt.show()

fig = plt.figure(figsize=(8,8))
plt.scatter(df.loc[df["file"].str.contains("OTf_c"), "mul_q"],df.loc[df["file"].str.contains("OTf_c"), "hir_q"], color="red", label="OTf")
plt.scatter(df.loc[df["file"].str.contains("SbF6_c"), "mul_q"],df.loc[df["file"].str.contains("SbF6_c"), "hir_q"], color="blue", label="SbF6")
plt.scatter(df.loc[df["file"].str.contains("dication"), "mul_q"],df.loc[df["file"].str.contains("dication"), "hir_q"], color="green", label="dication")
plt.legend(loc="best")
plt.ylabel("Mulliken Charge on Cu")
plt.xlabel("Hirshfeld Charge on Cu")
plt.title("Different Charge Metrics Are Consistent", fontsize="12", fontweight="bold")
plt.show()

fig = plt.figure(figsize=(8,8))
plt.scatter(df.loc[df["file"].str.contains("OTf_c"), "hir_q"],df.loc[df["file"].str.contains("OTf_c"), "dist"], color="red", label="OTf")
plt.scatter(df.loc[df["file"].str.contains("SbF6_c"), "hir_q"],df.loc[df["file"].str.contains("SbF6_c"), "dist"], color="blue", label="SbF6")
plt.legend(loc="best")
plt.ylabel("Cation–Anion Distance (Å)")
plt.xlabel("Hirshfeld Charge on Cu")
plt.title("SbF6 Results in More Cationic Contact Ion Pairs", fontsize="12", fontweight="bold")
plt.show()
```
